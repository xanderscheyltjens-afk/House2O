```python
import numpy as np
from numpy import deg2rad
```

```python
# chosen dimensions in m
length = 4 
depth = 4
height_full = 4
optimal_angle = deg2rad(38) # determined with optimalisation for solar energy absorption

# derived dimensions
sloped_window_height = np.tan(optimal_angle) * depth/2 # = vertical height of the roof alone
sloped_window_length = sloped_window_height / np.sin(optimal_angle)
height_walls = height_full-sloped_window_height

# volume
volume_cube = length*depth*height_walls
volume_roof = length*depth*sloped_window_height/2
volume = volume_cube + volume_roof

# surface area to collect power from the sun
vertical_window_area = height_walls * depth
roof_window_area = sloped_window_length * depth
total_window_area = vertical_window_area+roof_window_area

print(f"Hoogte van de rechte muren: {height_walls:.2f}m, hoogte van het dak: {sloped_window_height:.2f}m")
print(f"The volume of your house is {volume:.1f}m^3")
print(f"Total window surface area of your house: {total_window_area:.1f}m^2, of which {roof_window_area:.2f}m^2 is optimally sloped.")
print(f"Sloped window length: {sloped_window_length:.2f}")
```
